## whole image

```python
import time
import numpy as np
import cv2

path = r"D:\Dementia_Data_Preprocessing\testing_data\20211024\all\20211024_t055952.jpg"
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
arr = np.array(img)
arr2 = arr.flatten()

start = time.time()
a = np.std(arr2, ddof=1)
print("stdev:",a)
end = time.time()
print("cost time:",end-start)
```

## seperate to two areas

```python
import numpy as np
import cv2

path = r"D:\test6\magic_o\20211023_t073742.jpg"
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cropped = img[0:16, 0:32]
crop2 = img[16:24, 0:32]

arr = np.array(cropped)
arr2 = arr.flatten()

arr_b = np.array(crop2)
arr_b2 = arr_b.flatten()

a = np.std(arr2, ddof=1)
b = np.std(arr_b2, ddof=1)

print("bed stdev:",a)
print("floor stdev:",b)
```
