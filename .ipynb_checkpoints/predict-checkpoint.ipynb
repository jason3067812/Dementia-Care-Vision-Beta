```python
# only predict by model and save image

from PIL import Image
import numpy as np
from skimage import transform
import os
import shutil
import tensorflow as tf
from tensorflow import keras

def load(filename):
    
    np_image = Image.open(filename)
    np_image = np.array(np_image).astype('float32')/255
    np_image = transform.resize(np_image, (128,128, 3))
    np_image = np.expand_dims(np_image, axis=0)
    return np_image


model_path = r"C:\Users\ee527\Dementia_Vision_beta\model\test.h5"
directory_name = r"D:\Dementia_Data_Preprocessing\testing_data\all"
save = r"D:\Dementia_Data_Preprocessing\testing_data\test_exist\4"
save2 = r"D:\Dementia_Data_Preprocessing\testing_data\test_exist\0"

model = tf.keras.models.load_model(model_path)

num_count = 0
for file_name in os.listdir(directory_name):
    
    num_count = num_count+1
    
    image_path = directory_name + "/" + file_name  
    
    image = load(image_path)
    
    predict = model.predict(image)   
   
    
    if predict >= 0.5: 
        
        shutil.copyfile(image_path,save+"\\"+file_name)
    else:
        
        shutil.copyfile(image_path,save2+"\\"+file_name)
        
                
    
print("total number:", num_count)
```

```python
# check by bed data

from PIL import Image
import numpy as np
from skimage import transform
import os
import shutil
import tensorflow as tf
from tensorflow import keras

def load(filename):
    
    np_image = Image.open(filename)
    np_image = np.array(np_image).astype('float32')/255
    np_image = transform.resize(np_image, (128,128, 3))
    np_image = np.expand_dims(np_image, axis=0)
    return np_image


model_path = r"C:\Users\ee527\Dementia_Vision_beta\model\test.h5"
directory_name = r"D:\Dementia_Data_Preprocessing\testing_data\all"
save_error = False
error_path_1 = r"D:\Dementia_Data_Preprocessing\predict_error_data\denoise\0_to_4"
error_path_2 = r"D:\Dementia_Data_Preprocessing\predict_error_data\denoise\4_to_0"
ans = 4

model = tf.keras.models.load_model(model_path)

err_count = 0
num_count = 0
for file_name in os.listdir(directory_name):
    
    num_count = num_count+1
    
    image_path = directory_name + "/" + file_name  
    
    image = load(image_path)
    
    predict = model.predict(image)   
   
    if ans == 0:
        
        if predict >= 0.5:
        
            err_count = err_count+1
            print((image_path,predict))
            if save_error:
                shutil.copyfile(image_path,error_path_1+"\\"+file_name)
                
    elif ans == 4:
        
        if predict < 0.5:
        
            err_count = err_count+1
            print((image_path,predict))
            if save_error:
                shutil.copyfile(image_path,error_path_2+"\\"+file_name)
                
            
            
    #print((image_path,predict))
            
        
acc = ((num_count - err_count) / num_count) * 100

print("total number:", num_count)
print("error number:", err_count)
print(f"accuracy:{acc}%")
```
