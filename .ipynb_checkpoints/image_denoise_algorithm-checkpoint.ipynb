```python
import cv2
from collections import OrderedDict


img = cv2.imread(r'D:\test3\20210825_t022306.png',0)
img_copy = img
shape = img.shape
ymax = shape[0]
xmax = shape[1]

for i in range(ymax):
    
    for j in range(xmax):
        adjacent_list = []
        
        now_pixel = img[i][j]
        #print(f"點{(i,j)}的值:{now_pixel}")
        
        # 順序:上下左右       
        # 向上搜索
        if j-1 >= 0:
            up_pixel = img[i][j-1]
            adjacent_list.append(up_pixel)            
        
        # 向下搜索
        if j+1 < xmax:
            down_pixel = img[i][j+1]
            adjacent_list.append(down_pixel)
                        
        # 向左搜索
        if i-1 >= 0:
            left_pixel = img[i-1][j]
            adjacent_list.append(left_pixel)
            
        # 向右搜索
        if i+1 < ymax:
            right_pixel = img[i+1][j]
            adjacent_list.append(right_pixel)

        
        final_list = list(OrderedDict.fromkeys(adjacent_list))
    
        if len(final_list) == 1 and final_list[0] == 0:
            print("find noise")
            img_copy[i][j] = 0
            
cv2.imwrite("D:/test7/a.png", img_copy)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from collections import OrderedDict


# 修改路徑
directory_name = r'D:\Dementia_Data_Preprocessing\training_data\segmentation\0'
new_directory_name = r'D:\Dementia_Data_Preprocessing\training_data\denoise\0'

for filename in os.listdir(directory_name):
    

    base = filename.split(".")[0]
   
    # 讀取原始影像灰度顏色
    img = cv2.imread(directory_name + "/" + filename,0)
    
    
    img_copy = img
    shape = img.shape
    ymax = shape[0]
    xmax = shape[1]

    for i in range(ymax):

        for j in range(xmax):
            adjacent_list = []

            now_pixel = img[i][j]
            #print(f"點{(i,j)}的值:{now_pixel}")

            # 順序:上下左右       
            # 向上搜索
            if j-1 >= 0:
                up_pixel = img[i][j-1]
                adjacent_list.append(up_pixel)            

            # 向下搜索
            if j+1 < xmax:
                down_pixel = img[i][j+1]
                adjacent_list.append(down_pixel)

            # 向左搜索
            if i-1 >= 0:
                left_pixel = img[i-1][j]
                adjacent_list.append(left_pixel)

            # 向右搜索
            if i+1 < ymax:
                right_pixel = img[i+1][j]
                adjacent_list.append(right_pixel)


            final_list = list(OrderedDict.fromkeys(adjacent_list))

            if len(final_list) == 1 and final_list[0] == 0:
                img_copy[i][j] = 0

    cv2.imwrite(new_directory_name + "/" + base + ".png", img_copy)
```
