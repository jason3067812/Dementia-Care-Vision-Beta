```python
import os

directory_name = "C:/Users/jason/Desktop/datalogESP_pc/out"

delete_file = []
count = 0
for file_name in os.listdir(directory_name):
    
    count=count+1
    
    file = directory_name + "/" + file_name   
    print(file)
    
    file_size = os.path.getsize(file) 
    
    if file_size == 0:

        try:
            os.remove(file)
            
        except OSError as e:
            print(e)
            
        else:
            print("File is deleted successfully")
            delete_file.append(file)
            
    
print(" ")
print("original number:",count)
print("delete_file:",delete_file)
```
