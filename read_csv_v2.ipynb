```python
import pandas as pd
import csv
import numpy as np
from PIL import Image as im
```

```python
def to2sCompStr(num, bitWidth):
    num &= (2 << bitWidth-1)-1 # mask
    formatStr = '{:0'+str(bitWidth)+'b}'
    ret =  formatStr.format(int(num))
    return ret

def decode_data(decimal):
    
    
    data = to2sCompStr(decimal, 32)
    #print("data:",data)
    #print("length of data:",len(data))
    

    data1 = data[0:8]
    data2 = data[8:16]
    data3 = data[16:24]
    data4 = data[24:32]
    #print("before:",data1,data2,data3,data4)

    data1 = int(data1,2)
    data2 = int(data2,2)
    data3 = int(data3,2)
    data4 = int(data4,2)
    #print("after:",data1,data2,data3,data4)
    #print(" ")
    
    return data1, data2, data3, data4
```

```python

csv_path = "D:/sleep_detection_project/csv_data/data.csv"
save_img_path = "D:\\sleep_detection_project\\img\\data"

with open(csv_path, newline='') as csvfile:
    
    rows = list(csv.reader(csvfile))
    count = 0
    for row in rows[1:]:
        count+=1
        print(f"data: {count}, time: {row[0]}/{row[1]}, status: {row[2]}")
        
        date = row[0].split("/")
        new_date = f"{date[2]}_{date[0]}_{date[1]}"
        time = row[1].split(":")
        new_time = f"{time[0]}_{time[1]}_{time[2]}"

        img_name = f"{new_date}_{new_time}"
        
        path_name = save_img_path + "/" + f"{img_name}.jpg"
        
        data = row[3:-1]
        case = []
        for i in data:
    
            new_data = decode_data(int(i))
            case.append(new_data[3])
            case.append(new_data[2])
            case.append(new_data[1])
            case.append(new_data[0])
            
        arr = np.array(case)
        arr_2d = np.reshape(arr, (-1, 32))
        image = im.fromarray(arr_2d)
        image = image.convert("RGB")
        image.save(path_name)
        

    print("total data number:",count)
```
