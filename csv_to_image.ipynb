```python
import numpy as np
from numpy import genfromtxt
from PIL import Image
import os

directory_name = "D:/sleep_detection_project/decode_data/20210629"
save_directory = "D:/sleep_detection_project/image/20210629"

for file_name in os.listdir(directory_name):
    
    basename = os.path.basename(file_name)
    name = os.path.splitext(basename)[0]
    
    data = genfromtxt(directory_name + "/" + file_name, delimiter=',',dtype=None)
    data= data[:, :-1]
    data = data.astype(np.uint8)
    im = Image.fromarray(data)
    im = im.convert("RGB")
    im.save(save_directory + "/" + name + ".jpg")
    width, height = im.size
    print((width, height))

```
