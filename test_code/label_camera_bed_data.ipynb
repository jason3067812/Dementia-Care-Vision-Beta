```python
import os
import pandas as pd
import csv
import shutil

directory_name = r"D:\Dementia_Data_Preprocessing\main\label_image"

label_0_path = directory_name + "/" + "0"
label_1_path = directory_name + "/" + "1"
label_2_path = directory_name + "/" + "2"
label_3_path = directory_name + "/" + "3"
label_4_path = directory_name + "/" + "4"


if not os.path.exists(label_0_path):
    print(label_0_path)
    os.mkdir(label_0_path)
if not os.path.exists(label_1_path):
    print(label_1_path)
    os.mkdir(label_1_path)
if not os.path.exists(label_2_path):
    print(label_2_path)
    os.mkdir(label_2_path)
if not os.path.exists(label_3_path):
    print(label_3_path)
    os.mkdir(label_3_path)
if not os.path.exists(label_4_path):
    print(label_4_path)
    os.mkdir(label_4_path)
```

### read whisped csv file and decode image path

```python
csv_path = r"D:\Dementia_Data_Preprocessing\main\bed\20210816.csv"
image_path = r"D:\Dementia_Data_Preprocessing\main\out_image"

with open(csv_path, newline='') as csvfile:
    
    rows = list(csv.reader(csvfile))
 
    for row in rows[0:]:
        
        time = row[0]
        image_name = "t" + time[0] + time[1] + time[3] + time[4] + time[6] + time[7] + ".jpg"
        status = row[3]
 
        #print("time:",time," status:",status)
    
        for file_name in os.listdir(image_path):
            
  
            if image_name == file_name:
                
                print("same:",file_name)
                
                source = image_path + "/" + image_name
                
                if status == "00":
                    shutil.move(source, label_0_path)
                elif status == "01":
                    shutil.move(source, label_1_path)
                elif status == "02":
                    shutil.move(source, label_2_path)
                elif status == "03":
                    shutil.move(source, label_3_path)
                elif status == "04":
                    shutil.move(source, label_4_path)
                    
                break
                
            else:
                continue
    
       
```
