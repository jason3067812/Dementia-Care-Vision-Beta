```python
def to2sCompStr(num, bitWidth):
    num &= (2 << bitWidth-1)-1 # mask
    formatStr = '{:0'+str(bitWidth)+'b}'
    ret =  formatStr.format(int(num))
    return ret


def decode_data(decimal):
    
    
    data = to2sCompStr(decimal, 32)
#     print("data:",data)
#     print("length of data:",len(data))
#     print(" ")

    data1 = data[0:8]
    data2 = data[8:16]
    data3 = data[16:24]
    data4 = data[24:32]
    #print("before:",data1,data2,data3,data4)

    data1 = int(data1,2)
    data2 = int(data2,2)
    data3 = int(data3,2)
    data4 = int(data4,2)
    #print("after:",data1,data2,data3,data4)
    
    return data1, data2, data3, data4
```
